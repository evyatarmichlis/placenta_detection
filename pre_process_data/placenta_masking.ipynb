```python
import os
import json
import random
import shutil
import numpy as np
from pathlib import Path
import yaml
from tqdm import tqdm
import cv2
import matplotlib.pyplot as plt
from darwin.client import Client  # Ensure the Darwin SDK is installed
from shapely.geometry import Polygon as ShapelyPolygon, Point, MultiPolygon
from shapely.ops import unary_union

# === Configuration ===
API_KEY = '8QnldbG.FKUAeFkfHZAynvVRi9YkKIjlCFCUJ7_d'
DATASET_IDENTIFIER = 'extendedtrial/evyatar_project'
RAW_DIR = Path("/content/raw")
PROCESSED_DIR = Path("/content/processed2")
TRAIN_SPLIT = 0.7
VAL_SPLIT = 0.15
TEST_SPLIT = 0.15

# === Constants for filtering ===
# Minimum size for the final placenta segmentation (as % of image area)
MIN_PLACENTA_SIZE = 0.15  # 15% of image area
# Minimum size for individual polygons to be considered (as % of image area)
MIN_POLYGON_SIZE = 0.08   # 8% of image area
# Maximum distance from image center for valid placenta (as % of diagonal)
MAX_CENTER_DISTANCE = 0.35  # 35% of image diagonal

# === Optional: Download the dataset from Darwin and ensure it's in the right location ===
def download_dataset(api_key: str, dataset_id: str, raw_dir: Path):
    """
    Download dataset from Darwin and ensure it's in the RAW_DIR location.
    """
    client = Client.from_api_key(api_key)
    dataset = client.get_remote_dataset(dataset_id)

    # Use the specific paths that we now know work
    darwin_images_path = Path("/root/.darwin/datasets/extendedtrial/evyatar_project/images")
    darwin_annotations_path = Path("/root/.darwin/datasets/extendedtrial/evyatar_project/releases/complete/annotations")

    # Ensure paths exist
    if not darwin_images_path.exists():
        # Try using the Darwin API to get the path
        try:
            dataset.pull()  # Pull the dataset
            print("Dataset downloaded using Darwin SDK.")
        except Exception as e:
            print(f"Error downloading dataset: {e}")
            print("Please download the dataset manually and place files in:")
            print(f" - Images: {raw_dir}/images")
            print(f" - Annotations: {raw_dir}/annotations")
            return 0, 0

    # Create target directories
    raw_images_dir = raw_dir / "images"
    raw_annotations_dir = raw_dir / "annotations"
    raw_images_dir.mkdir(parents=True, exist_ok=True)
    raw_annotations_dir.mkdir(parents=True, exist_ok=True)

    # Copy images
    images_copied = 0
    if darwin_images_path.exists():
        print(f"Copying images from {darwin_images_path}")
        for img_path in darwin_images_path.glob("*.jpg"):
            dest_path = raw_images_dir / img_path.name
            shutil.copy2(img_path, dest_path)
            images_copied += 1
            if images_copied <= 5:  # Show the first 5 files
                print(f"Copied image: {img_path.name}")
        print(f"Copied {images_copied} images")
    else:
        print(f"Warning: Images directory not found at {darwin_images_path}")

    # Copy annotations
    annotations_copied = 0
    if darwin_annotations_path.exists():
        print(f"Copying annotations from {darwin_annotations_path}")
        for ann_path in darwin_annotations_path.glob("*.json"):
            dest_path = raw_annotations_dir / ann_path.name
            shutil.copy2(ann_path, dest_path)
            annotations_copied += 1
            if annotations_copied <= 5:  # Show the first 5 files
                print(f"Copied annotation: {ann_path.name}")
        print(f"Copied {annotations_copied} annotations")
    else:
        print(f"Warning: Annotations directory not found at {darwin_annotations_path}")

    # Final verification
    images = list(raw_images_dir.glob("*.jpg"))
    annotations = list(raw_annotations_dir.glob("*.json"))

    print(f"Raw directory now contains {len(images)} images and {len(annotations)} annotations")

    return len(images), len(annotations)

# === Enhanced conversion function for annotations to YOLO segmentation format ===
def convert_annotation_to_yolo(annotation_path: Path, img_path=None, save_mask=True, masks_dir=None, debug=False):
    """
    Reads a Darwin JSON file and converts the placenta polygon to YOLO segmentation format.
    Steps:
      1. Filter small or invalid polygons
      2. Find the main placenta polygon (large and centrally located)
      3. Merge nearby polygons if needed
      4. Compute the convex hull and ensure it meets size requirements
      5. Generate YOLO format line and visualization

    Args:
        annotation_path: Path to the annotation JSON file
        img_path: Path to the corresponding image (for mask visualization)
        save_mask: Whether to save a visualization mask
        masks_dir: Directory to save masks
        debug: Enable debug output

    Returns:
        YOLO format segmentation line
    """
    with open(annotation_path, 'r') as f:
        ann = json.load(f)

    # Get image dimensions from the first slot.
    slot = ann["item"]["slots"][0]
    width = float(slot["width"])
    height = float(slot["height"])
    image_center = Point(width / 2, height / 2)
    image_area = width * height
    max_distance = np.sqrt(width**2 + height**2) / 2  # Half diagonal

    # Convert percentage thresholds to actual values
    min_placenta_area = image_area * MIN_PLACENTA_SIZE
    min_polygon_area = image_area * MIN_POLYGON_SIZE
    max_center_dist = max_distance * MAX_CENTER_DISTANCE

    # Initialize counters
    small_polygons = 0
    invalid_polygons = 0
    good_polygons = 0

    # Gather all candidate polygons (absolute coordinates)
    candidate_polygons = []

    for annotation in ann.get("annotations", []):
        annotation_name = annotation.get("name", "").lower()
        annotation_class = annotation.get("class", {}).get("name", "").lower()

        # Optional filter for annotations with specific names
        # is_placenta = "placenta" in annotation_name or "placenta" in annotation_class

        if "polygon" in annotation and "paths" in annotation["polygon"]:
            for path_idx, path in enumerate(annotation["polygon"]["paths"]):
                if len(path) >= 3:  # Need at least 3 points for a polygon
                    try:
                        pts = [(float(pt["x"]), float(pt["y"])) for pt in path]
                        poly = ShapelyPolygon(pts)

                        # Check if the polygon is valid
                        if not poly.is_valid:
                            if debug:
                                print(f"Invalid polygon found in {annotation_path.name}, path #{path_idx+1}")
                            invalid_polygons += 1
                            continue

                        # Calculate area and percentage
                        area = poly.area
                        area_percentage = (area / image_area) * 100

                        # Check if the polygon is large enough
                        if area < min_polygon_area:
                            if debug:
                                print(f"Small polygon in {annotation_path.name}, path #{path_idx+1} "
                                      f"(area: {area:.1f} px, {area_percentage:.2f}% of image)")
                            small_polygons += 1
                            continue

                        # Check if the polygon is near the center of the image
                        centroid = poly.centroid
                        dist_to_center = centroid.distance(image_center)
                        if dist_to_center > max_center_dist:
                            if debug:
                                print(f"Polygon too far from center in {annotation_path.name}, "
                                      f"distance: {dist_to_center:.1f} px, "
                                      f"{(dist_to_center/max_distance)*100:.1f}% of max diagonal")
                            continue

                        # This polygon passes initial filtering
                        candidate_polygons.append(poly)
                        good_polygons += 1

                    except Exception as e:
                        if debug:
                            print(f"Error creating polygon in {annotation_path.name}, path #{path_idx+1}: {e}")
                        invalid_polygons += 1

    # Report statistics if debugging
    if debug:
        print(f"File {annotation_path.name}: {good_polygons} good polygons, {small_polygons} small polygons, "
              f"{invalid_polygons} invalid polygons")

    if not candidate_polygons:
        raise ValueError(f"No valid placenta polygons found in {annotation_path.name} "
                         f"(min size: {MIN_PLACENTA_SIZE*100}% of image)")

    # Find the best placenta polygon - scored by size and proximity to center
    best_poly = None
    best_score = float('-inf')

    for poly in candidate_polygons:
        # Calculate normalized distance (1 = at center, 0 = at max distance)
        centroid = poly.centroid
        dist = centroid.distance(image_center)
        norm_dist = 1 - (dist / max_distance)

        # Calculate normalized area (relative to image size)
        area_ratio = poly.area / image_area

        # Score combines proximity to center and size (weight size more)
        score = 0.4 * norm_dist + 0.6 * area_ratio

        if debug:
            print(f"Polygon score: {score:.3f} (area: {area_ratio*100:.1f}%, center dist: {dist:.1f}px)")

        if score > best_score:
            best_score = score
            best_poly = poly

    # Try to merge nearby polygons with the best one
    if len(candidate_polygons) > 1:
        # Create buffer to help join nearly touching polygons
        buffer_size = width * 0.03  # 3% of image width
        merged_poly = best_poly.buffer(buffer_size)

        # Find and merge nearby polygons
        merge_count = 0
        for poly in candidate_polygons:
            if poly != best_poly and poly.distance(best_poly) < width * 0.1:  # 10% of width
                merged_poly = unary_union([merged_poly, poly])
                merge_count += 1

        if merge_count > 0 and debug:
            print(f"Merged {merge_count} additional polygons with the best one")

        # Remove the buffer we added
        merged_poly = merged_poly.buffer(-buffer_size)
        if merged_poly.is_empty:
            # Fall back to original if buffer removal caused issues
            merged_poly = best_poly
    else:
        merged_poly = best_poly

    # Final validation of the merged polygon
    if merged_poly.area < min_placenta_area:
        raise ValueError(f"Final placenta polygon in {annotation_path.name} is too small: "
                         f"{merged_poly.area/image_area*100:.1f}% of image (min: {MIN_PLACENTA_SIZE*100}%)")

    # Get the convex hull
    # In convert_annotation_to_yolo
    # After creating the hull_poly
    hull_poly = merged_poly.convex_hull

    # Add an extra check to ensure perfect convexity
    hull_coords = np.array(hull_poly.exterior.coords)[:-1]  # Remove duplicate endpoint
    is_convex = True
    for i in range(len(hull_coords)):
        p1 = hull_coords[i-2]
        p2 = hull_coords[i-1]
        p3 = hull_coords[i]
        # Calculate cross product to check convexity
        cross_product = (p2[0] - p1[0]) * (p3[1] - p2[1]) - (p2[1] - p1[1]) * (p3[0] - p2[0])
        if cross_product < 0:  # Not convex
            is_convex = False
            break

    if not is_convex:
        print(f"Warning: Hull for {annotation_path.name} is not perfectly convex. Enforcing convexity.")
        # Force convexity using scikit-learn's ConvexHull
        from scipy.spatial import ConvexHull
        points = np.array(list(hull_poly.exterior.coords)[:-1])
        hull = ConvexHull(points)
        hull_coords = points[hull.vertices]

    # Check the hull size again
    if hull_poly.area < min_placenta_area:
        raise ValueError(f"Convex hull in {annotation_path.name} is too small: "
                         f"{hull_poly.area/image_area*100:.1f}% of image (min: {MIN_PLACENTA_SIZE*100}%)")

    # Simplify to reduce the number of points while preserving shape
    # Use a tolerance relative to image dimensions
    tolerance = max(width, height) * 0.005  # 0.5% of the largest dimension
    simplified_hull = hull_poly.simplify(tolerance, preserve_topology=True)

    # Get coordinates of the simplified hull
    hull_coords = list(simplified_hull.exterior.coords)[:-1]  # Remove duplicate endpoint

    # Compute bounding box
    minx, miny, maxx, maxy = simplified_hull.bounds
    bbox_center_x = (minx + maxx) / 2
    bbox_center_y = (miny + maxy) / 2
    bbox_width = maxx - minx
    bbox_height = maxy - miny

    # Check bounding box relative to image size
    if bbox_width < width * 0.15 or bbox_height < height * 0.15:
        raise ValueError(f"Bounding box in {annotation_path.name} is too small: "
                        f"{bbox_width/width*100:.1f}% × {bbox_height/height*100:.1f}% of image size")

    # Normalize values
    norm_bbox_center_x = bbox_center_x / width
    norm_bbox_center_y = bbox_center_y / height
    norm_bbox_width = bbox_width / width
    norm_bbox_height = bbox_height / height

    # Normalize hull coordinates
    normalized_hull = []
    for x, y in hull_coords:
        normalized_hull.append(f"{x/width:.6f}")
        normalized_hull.append(f"{y/height:.6f}")

    # Format YOLO line: class_id bbox_center_x bbox_center_y bbox_width bbox_height x1 y1 x2 y2...
    yolo_line = f"0 {norm_bbox_center_x:.6f} {norm_bbox_center_y:.6f} {norm_bbox_width:.6f} {norm_bbox_height:.6f} " + " ".join(normalized_hull)

    # Save visualization mask if requested
    if save_mask and masks_dir is not None:
        # Create a blank mask
        mask = np.zeros((int(height), int(width), 3), dtype=np.uint8)

        # Get hull points for drawing
        hull_points = np.array(hull_coords, dtype=np.int32)

        # Draw the convex hull on the mask (white filled polygon)
        cv2.fillPoly(mask, [hull_points], (255, 255, 255))

        # Draw the convex hull boundary in red
        cv2.polylines(mask, [hull_points], True, (0, 0, 255), 2)

        # Save the mask
        mask_path = masks_dir / f"{img_path.stem}_mask.jpg"
        cv2.imwrite(str(mask_path), mask)

        # If we have the original image, create a side-by-side comparison
        if img_path is not None and img_path.exists():
            try:
                # Read the original image
                img = cv2.imread(str(img_path))
                if img is not None:
                    # Resize if dimensions don't match
                    if img.shape[0] != int(height) or img.shape[1] != int(width):
                        img = cv2.resize(img, (int(width), int(height)))

                    # Create overlay image (avoid using alpha parameter)
                    green_mask = np.zeros_like(img, dtype=np.uint8)
                    cv2.fillPoly(green_mask, [hull_points], (0, 255, 0))
                    img_with_overlay = cv2.addWeighted(img, 0.7, green_mask, 0.3, 0)
                    cv2.polylines(img_with_overlay, [hull_points], True, (0, 0, 255), 2)

                    # Add hull area percentage to the image
                    area_percent = hull_poly.area / image_area * 100
                    text = f"Area: {area_percent:.1f}% of image"
                    cv2.putText(img_with_overlay, text, (20, 30),
                                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)

                    # Create a side-by-side comparison
                    comparison = np.hstack((img, img_with_overlay))

                    # Save the comparison
                    comparison_path = masks_dir / f"{img_path.stem}_comparison.jpg"
                    cv2.imwrite(str(comparison_path), comparison)
            except Exception as e:
                if debug:
                    print(f"Error creating image overlay: {e}")

    if debug:
        area_percentage = hull_poly.area / image_area * 100
        print(f"Processed {annotation_path.name}: {len(hull_coords)} hull points, "
              f"area: {area_percentage:.1f}% of image (valid)")

    return yolo_line

# === Create YOLO dataset structure (train/val/test splits) ===
def create_yolo_dataset_structure(raw_dir: Path, processed_dir: Path, debug=False):
    images_dir = raw_dir / "images"
    ann_dir = raw_dir / "annotations"

    # Create processed directories for each split (images & labels)
    for split in ["train", "val", "test"]:
        (processed_dir / split / "images").mkdir(parents=True, exist_ok=True)
        (processed_dir / split / "labels").mkdir(parents=True, exist_ok=True)
        # Create masks directory for visualization
        (processed_dir / split / "masks").mkdir(parents=True, exist_ok=True)

    # Gather annotation files and match them with images.
    ann_files = list(ann_dir.glob("*.json"))
    print(f"Found {len(ann_files)} annotation files")

    data = []
    for ann_file in ann_files:
        with open(ann_file, 'r') as f:
            ann = json.load(f)
        img_name = ann['item']['name']
        img_path = images_dir / img_name
        if img_path.exists():
            data.append((img_path, ann_file))
        else:
            print(f"Warning: Image file {img_name} not found.")

    print(f"Matched {len(data)} images with annotations")

    # Shuffle data and split into train, val, test.
    random.seed(42)  # For reproducibility
    random.shuffle(data)
    n_total = len(data)
    n_train = int(n_total * TRAIN_SPLIT)
    n_val = int(n_total * VAL_SPLIT)
    train_data = data[:n_train]
    val_data = data[n_train:n_train + n_val]
    test_data = data[n_train + n_val:]

    # Keep track of success/failure counts
    processed_counts = {"train": 0, "val": 0, "test": 0}
    too_small_counts = {"train": 0, "val": 0, "test": 0}
    other_error_counts = {"train": 0, "val": 0, "test": 0}

    # Process each split.
    for split_name, split_data in zip(["train", "val", "test"], [train_data, val_data, test_data]):
        print(f"\nProcessing {split_name} split with {len(split_data)} images...")
        masks_dir = processed_dir / split_name / "masks"

        for img_path, ann_file in tqdm(split_data, desc=f"{split_name} split"):
            try:
                # Generate YOLO label with mask visualization
                yolo_label = convert_annotation_to_yolo(
                    annotation_path=ann_file,
                    img_path=img_path,
                    save_mask=True,
                    masks_dir=masks_dir,
                    debug=debug
                )

                # Copy image file
                dest_img = processed_dir / split_name / "images" / img_path.name
                shutil.copy2(img_path, dest_img)

                # Save YOLO label file
                label_filename = img_path.stem + ".txt"
                dest_label = processed_dir / split_name / "labels" / label_filename
                with open(dest_label, 'w') as f:
                    f.write(yolo_label)

                processed_counts[split_name] += 1

            except ValueError as e:
                error_message = str(e).lower()
                # Categorize the error
                if "small" in error_message or "too small" in error_message:
                    if debug:
                        print(f"Skipping {img_path.name}: {e}")
                    too_small_counts[split_name] += 1
                else:
                    if debug:
                        print(f"Error processing {img_path.name}: {e}")
                    other_error_counts[split_name] += 1
            except Exception as e:
                if debug:
                    print(f"Unexpected error processing {img_path.name}: {e}")
                other_error_counts[split_name] += 1

    # Report processing summary
    print("\nProcessing Summary:")
    for split in ["train", "val", "test"]:
        success = processed_counts[split]
        too_small = too_small_counts[split]
        other_errors = other_error_counts[split]
        total = success + too_small + other_errors
        success_rate = (success / total * 100) if total > 0 else 0
        print(f"{split}: {success}/{total} successful ({success_rate:.1f}%)")
        print(f"  - {too_small} images skipped (placenta too small)")
        print(f"  - {other_errors} images failed for other reasons")

# === Create YAML file for YOLO training configuration ===
def create_yaml_file(processed_dir: Path, yaml_path: Path):
    data_yaml = {
        "path": str(processed_dir.resolve()),
        "train": str((processed_dir / "train" / "images").resolve()),
        "val": str((processed_dir / "val" / "images").resolve()),
        "test": str((processed_dir / "test" / "images").resolve()),
        "nc": 1,
        "names": ["placenta"]
    }
    with open(yaml_path, 'w') as f:
        yaml.dump(data_yaml, f, default_flow_style=False)
    print(f"YAML configuration created at: {yaml_path}")

# === Verify the dataset format ===
def verify_dataset(processed_dir: Path):
    total_images = 0
    total_labels = 0

    for split in ["train", "val", "test"]:
        img_dir = processed_dir / split / "images"
        label_dir = processed_dir / split / "labels"
        if not img_dir.exists() or not label_dir.exists():
            print(f"Warning: {split} directories not found!")
            continue
        images = list(img_dir.glob("*.*"))
        labels = list(label_dir.glob("*.txt"))
        total_images += len(images)
        total_labels += len(labels)
        print(f"{split}: {len(images)} images, {len(labels)} label files")
    print(f"Total: {total_images} images, {total_labels} label files")

# === Clean existing dataset ===
def clean_dataset(processed_dir: Path):
    if processed_dir.exists():
        for split in ["train", "val", "test"]:
            split_dir = processed_dir / split
            if split_dir.exists():
                print(f"Cleaning {split} directory...")
                shutil.rmtree(split_dir)
        yaml_file = processed_dir / "data.yaml"
        if yaml_file.exists():
            os.remove(yaml_file)
    print("Cleaned existing processed data")


def verify_yolo_format_masks(processed_dir: Path, num_samples=5):
    """Verify that YOLO format labels match expected convex hulls"""
    for split in ["train", "val", "test"]:
        # Get random samples
        img_dir = processed_dir / split / "images"
        label_dir = processed_dir / split / "labels"

        image_files = list(img_dir.glob("*.jpg"))
        if len(image_files) <= num_samples:
            samples = image_files
        else:
            samples = random.sample(image_files, num_samples)

        for img_path in samples:
            # Load image
            img = cv2.imread(str(img_path))
            h, w = img.shape[:2]

            # Load corresponding label
            label_path = label_dir / f"{img_path.stem}.txt"
            if not label_path.exists():
                continue

            # Parse YOLO format file
            with open(label_path, 'r') as f:
                label_data = f.read().strip().split()

            if len(label_data) < 5:
                continue

            # Extract polygon coordinates
            points = []
            for i in range(5, len(label_data), 2):
                x = float(label_data[i]) * w
                y = float(label_data[i+1]) * h
                points.append((int(x), int(y)))

            # Draw the polygon on the image
            overlay = img.copy()
            if points:
                points_array = np.array(points, dtype=np.int32)
                cv2.fillPoly(overlay, [points_array], (0, 255, 0), 8)

            # Create comparison
            verification = np.hstack((img, cv2.addWeighted(img, 0.7, overlay, 0.3, 0)))

            # Save verification image
            verify_dir = processed_dir / "verification"
            verify_dir.mkdir(exist_ok=True)
            cv2.imwrite(str(verify_dir / f"{split}_{img_path.stem}_yolo_verification.jpg"), verification)

# === Create a helper function to visualize masks ===
def visualize_random_samples(processed_dir: Path, num_samples=5):
    """
    Visualize random samples from each split to help verify the segmentation quality.

    Args:
        processed_dir: Path to the processed directory
        num_samples: Number of random samples to visualize from each split
    """
    plt.figure(figsize=(15, 5*num_samples))

    row_idx = 0
    for split in ["train", "val", "test"]:
        # Get all mask comparison files
        mask_dir = processed_dir / split / "masks"
        if not mask_dir.exists():
            print(f"No masks found for {split} split")
            continue

        comparison_files = list(mask_dir.glob("*_comparison.jpg"))
        if not comparison_files:
            print(f"No comparison files found for {split} split")
            continue

        # Select random samples
        if len(comparison_files) <= num_samples:
            samples = comparison_files
        else:
            samples = random.sample(comparison_files, num_samples)

        # Sort by file size to prioritize interesting examples
        samples.sort(key=lambda f: f.stat().st_size, reverse=True)

        # Plot each sample
        for i, sample_path in enumerate(samples):
            try:
                img = cv2.imread(str(sample_path))
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert from BGR to RGB

                row_idx += 1
                plt.subplot(num_samples * 3, 1, row_idx)
                plt.imshow(img)
                plt.title(f"{split} - {sample_path.stem}")
                plt.axis('off')
            except Exception as e:
                print(f"Error visualizing {sample_path}: {e}")

    plt.tight_layout()
    vis_path = processed_dir / "segmentation_samples.png"
    plt.savefig(vis_path)
    print(f"Visualization saved to {vis_path}")

# === Main execution ===
if __name__ == "__main__":
    # Enable debug mode for more verbose output
    DEBUG = True

    print("=== Placenta Segmentation Dataset Preparation ===")
    print(f"Minimum placenta size: {MIN_PLACENTA_SIZE*100:.1f}% of image area")
    print(f"Minimum polygon size: {MIN_POLYGON_SIZE*100:.1f}% of image area")
    print(f"Maximum center distance: {MAX_CENTER_DISTANCE*100:.1f}% of image diagonal")

    # Download the dataset from Darwin and ensure it's in the right location
    try:
        images_count, annotations_count = download_dataset(API_KEY, DATASET_IDENTIFIER, RAW_DIR)
        print(f"Dataset ready for processing with {images_count} images and {annotations_count} annotations")
    except Exception as e:
        print(f"Error during dataset download: {e}")
        print("Continuing with existing files if any...")

    # Clean existing processed data.
    clean_dataset(PROCESSED_DIR)
    PROCESSED_DIR.mkdir(parents=True, exist_ok=True)

    # Create YOLO dataset structure and convert annotations.
    create_yolo_dataset_structure(RAW_DIR, PROCESSED_DIR, debug=DEBUG)

    # Create YAML configuration file.
    create_yaml_file(PROCESSED_DIR, PROCESSED_DIR / "data.yaml")

    # Verify dataset format.
    verify_dataset(PROCESSED_DIR)
    verify_yolo_format_masks(PROCESSED_DIR)
    # Visualize some random samples to check segmentation quality
    try:
        visualize_random_samples(PROCESSED_DIR, num_samples=5)
    except Exception as e:
        print(f"Error generating visualization: {e}")

    print("\nDataset preparation complete!")
    print("Training command: yolo segment train data=/content/processed/data.yaml model=yolov8n-seg.pt epochs=50")
    print("\nTip: Check the 'masks' directory in each split to verify the segmentation quality.")
```

```python
import cv2
import numpy as np
from pathlib import Path
from ultralytics import YOLO
from tqdm import tqdm
import torch
import random
from google.colab import drive
drive.mount('/content/drive')

class MaskProcessor:
    def __init__(self, model_path):
        self.model = YOLO(model_path)
        self.device = 'cuda' if torch.cuda.is_available() else 'cpu'
        print(f"Using device: {self.device}")

    def create_masked_image(self, image, masks,
                            mask_mode='ultra_convex',
                            buffer_size=5,
                            force_ellipse=True,      # Force ellipse fitting
                            pre_closing_size=25,     # Very aggressive morphological closing
                            convexity_iterations=2,  # Multiple convexity passes
                            min_contour_area_percent=0.01,
                            save_intermediate=True,
                            output_dir=None,
                            img_name=None):
        """
        Create an ultra-convex masked image with guaranteed no cuts.

        Parameters:
            image: Input image
            masks: Model masks
            mask_mode: 'ultra_convex', 'ellipse', 'strict_convex_hull'
            buffer_size: Size of buffer to add around mask (px)
            force_ellipse: Whether to fit an ellipse to ensure convexity
            pre_closing_size: Size of kernel for aggressive pre-processing
            convexity_iterations: Number of convexity passes to apply
            min_contour_area_percent: Min contour area as % of image size
            save_intermediate: Save intermediate processing steps
            output_dir: Directory to save intermediate images
            img_name: Base name for saving intermediate images
        """
        h, w = image.shape[:2]
        image_area = h * w
        min_contour_area = image_area * min_contour_area_percent

        # Create output for debugging
        if save_intermediate and output_dir and img_name:
            debug_dir = Path(output_dir) / "debug"
            debug_dir.mkdir(parents=True, exist_ok=True)

        # Initialize the output mask
        masked_image = np.zeros_like(image)

        # 1) Combine all model masks
        if masks is None or len(masks) == 0:
            return masked_image

        combined = np.zeros((h, w), dtype=bool)
        for m in masks:
            m_resized = cv2.resize(m.astype(float), (w, h))
            combined |= (m_resized > 0.5)

        mask_uint8 = (combined * 255).astype(np.uint8)

        # Save original mask for debugging
        if save_intermediate:
            debug_img = np.zeros_like(image)
            debug_img[combined] = (0, 0, 255)  # Red mask
            cv2.imwrite(str(debug_dir / f"{img_name}_1_raw_mask.jpg"), debug_img)

        # 2) ULTRA-AGGRESSIVE pre-processing to ensure convexity
        # Apply very large morphological closing to fill even large internal gaps
        if pre_closing_size > 0:
            close_kernel = cv2.getStructuringElement(
                cv2.MORPH_ELLIPSE,
                (pre_closing_size, pre_closing_size)
            )
            mask_closed = cv2.morphologyEx(
                mask_uint8,
                cv2.MORPH_CLOSE,
                close_kernel,
                iterations=3  # Multiple iterations for deeper filling
            )

            # Save closed mask for debugging
            if save_intermediate:
                debug_img = np.zeros_like(image)
                debug_img[mask_closed > 0] = (255, 0, 0)  # Blue mask
                cv2.imwrite(str(debug_dir / f"{img_name}_2_closed_mask.jpg"), debug_img)
        else:
            mask_closed = mask_uint8

        # 3) Find contours on the pre-processed mask
        contours, _ = cv2.findContours(
            mask_closed,
            cv2.RETR_EXTERNAL,
            cv2.CHAIN_APPROX_SIMPLE
        )

        # Filter small contours (noise)
        filtered_contours = [cnt for cnt in contours if cv2.contourArea(cnt) >= min_contour_area]

        # No valid contours found
        if not filtered_contours:
            return masked_image

        # 4) Process based on selected mode
        if mask_mode == 'ultra_convex':
            # Find the largest contour
            largest_contour = max(filtered_contours, key=cv2.contourArea)

            # Start with the original contour
            processed_contour = largest_contour

            # Apply multiple passes of convex hull for ultra-aggressive convexity
            for i in range(convexity_iterations):
                # Get convex hull
                processed_contour = cv2.convexHull(processed_contour)

                # Save intermediate hulls for debugging
                if save_intermediate:
                    debug_img = np.zeros_like(image)
                    cv2.drawContours(debug_img, [processed_contour], -1, (0, 255, 0), thickness=-1)
                    cv2.imwrite(str(debug_dir / f"{img_name}_3_{i+1}_hull_pass.jpg"), debug_img)

            # Create mask from the final contour
            final_mask = np.zeros_like(mask_uint8)
            cv2.drawContours(final_mask, [processed_contour], -1, 255, thickness=-1)

            # Optional: Force ellipse fitting for guaranteed convexity
            if force_ellipse:
                # Create an ellipse mask
                ellipse_mask = np.zeros_like(mask_uint8)

                # Fit an ellipse to the contour
                if len(processed_contour) >= 5:  # Need at least 5 points for ellipse fitting
                    ellipse = cv2.fitEllipse(processed_contour)
                    cv2.ellipse(ellipse_mask, ellipse, 255, thickness=-1)

                    # Save ellipse for debugging
                    if save_intermediate:
                        debug_img = np.zeros_like(image)
                        cv2.ellipse(debug_img, ellipse, (0, 255, 255), thickness=-1)
                        cv2.imwrite(str(debug_dir / f"{img_name}_4_ellipse.jpg"), debug_img)

                    # Combine ellipse with convex hull (take maximum of both)
                    final_mask = np.maximum(final_mask, ellipse_mask)

            # Add buffer if needed
            if buffer_size > 0:
                kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (buffer_size, buffer_size))
                final_mask = cv2.dilate(final_mask, kernel, iterations=1)

                # Save buffered mask for debugging
                if save_intermediate:
                    debug_img = np.zeros_like(image)
                    buffered_contours, _ = cv2.findContours(final_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
                    cv2.drawContours(debug_img, buffered_contours, -1, (0, 255, 255), thickness=-1)
                    cv2.imwrite(str(debug_dir / f"{img_name}_5_final_mask.jpg"), debug_img)

            # Apply mask
            mask_bool = final_mask.astype(bool)
            masked_image = np.where(mask_bool[:, :, None], image, 0)

        elif mask_mode == 'ellipse':
            # Find the largest contour
            largest_contour = max(filtered_contours, key=cv2.contourArea)

            # Create an ellipse mask (guaranteed to be convex)
            ellipse_mask = np.zeros_like(mask_uint8)

            # Fit an ellipse to the contour if possible
            if len(largest_contour) >= 5:  # Need at least 5 points for ellipse fitting
                ellipse = cv2.fitEllipse(largest_contour)
                cv2.ellipse(ellipse_mask, ellipse, 255, thickness=-1)

                # Add buffer if needed
                if buffer_size > 0:
                    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (buffer_size, buffer_size))
                    ellipse_mask = cv2.dilate(ellipse_mask, kernel, iterations=1)

                # Apply mask
                mask_bool = ellipse_mask.astype(bool)
                masked_image = np.where(mask_bool[:, :, None], image, 0)
            else:
                # Fallback to standard convex hull if ellipse fitting fails
                hull = cv2.convexHull(largest_contour)

                # Create hull mask
                hull_mask = np.zeros_like(mask_uint8)
                cv2.drawContours(hull_mask, [hull], -1, 255, thickness=-1)

                # Add buffer if needed
                if buffer_size > 0:
                    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (buffer_size, buffer_size))
                    hull_mask = cv2.dilate(hull_mask, kernel, iterations=1)

                # Apply mask
                mask_bool = hull_mask.astype(bool)
                masked_image = np.where(mask_bool[:, :, None], image, 0)

        elif mask_mode == 'strict_convex_hull':
            # Find the largest contour
            largest_contour = max(filtered_contours, key=cv2.contourArea)

            # Get the convex hull
            hull = cv2.convexHull(largest_contour)

            # Create hull mask
            hull_mask = np.zeros_like(mask_uint8)
            cv2.drawContours(hull_mask, [hull], -1, 255, thickness=-1)

            # Add buffer if needed
            if buffer_size > 0:
                kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (buffer_size, buffer_size))
                hull_mask = cv2.dilate(hull_mask, kernel, iterations=1)

            # Apply mask
            mask_bool = hull_mask.astype(bool)
            masked_image = np.where(mask_bool[:, :, None], image, 0)

        return masked_image

    def process_dataset(self, input_dir, output_dir,
                        mask_mode='ultra_convex',
                        buffer_size=5,
                        force_ellipse=True,
                        pre_closing_size=25,
                        convexity_iterations=2,
                        min_contour_area_percent=0.01,
                        save_intermediate=True):
        """Process the entire dataset with ultra-convex masks and save results.

        Parameters:
          input_dir (str): Directory containing input images.
          output_dir (str): Directory to store predictions and masked images.
          mask_mode (str): 'ultra_convex', 'ellipse', or 'strict_convex_hull'.
          buffer_size (int): Buffer size to apply to the mask.
          force_ellipse (bool): Whether to force ellipse fitting for guaranteed convexity.
          pre_closing_size (int): Size of kernel for aggressive pre-processing.
          convexity_iterations (int): Number of convexity passes to apply.
          min_contour_area_percent (float): Minimum contour area as % of image size.
          save_intermediate (bool): Save intermediate processing steps.
        """
        input_dir = Path(input_dir)
        output_dir = Path(output_dir)

        # Create output directories
        predictions_dir = output_dir / 'predictions'
        masked_dir = output_dir / 'masked_images'
        predictions_dir.mkdir(parents=True, exist_ok=True)
        masked_dir.mkdir(parents=True, exist_ok=True)

        # Get all image files
        image_files = list(input_dir.glob('*.jpg')) + list(input_dir.glob('*.png'))
        print(f"Found {len(image_files)} images to process")

        # Print processing parameters
        print(f"\nProcessing parameters:")
        print(f"  Mode: {mask_mode}")
        print(f"  Buffer size: {buffer_size}px")
        print(f"  Pre-closing size: {pre_closing_size}px")
        print(f"  Force ellipse fitting: {force_ellipse}")
        print(f"  Convexity iterations: {convexity_iterations}")
        print(f"  Min contour area: {min_contour_area_percent*100}% of image\n")

        for img_path in tqdm(image_files, desc="Processing images"):
            try:
                image = cv2.imread(str(img_path))
                if image is None:
                    print(f"Failed to read image: {img_path}")
                    continue

                # Get predictions from the model
                results = self.model.predict(
                    source=image,
                    conf=0.25,  # Confidence threshold
                    save=True,
                    project=str(predictions_dir),
                    name='',
                    exist_ok=True
                )

                # Process the mask from the first result (if available)
                if len(results) > 0 and hasattr(results[0], 'masks') and results[0].masks is not None:
                    masks = results[0].masks.data.cpu().numpy()
                    try:
                        masked_image = self.create_masked_image(
                            image,
                            masks,
                            mask_mode=mask_mode,
                            buffer_size=buffer_size,
                            force_ellipse=force_ellipse,
                            pre_closing_size=pre_closing_size,
                            convexity_iterations=convexity_iterations,
                            min_contour_area_percent=min_contour_area_percent,
                            save_intermediate=save_intermediate,
                            output_dir=output_dir,
                            img_name=img_path.stem
                        )
                        output_path = masked_dir / f"{img_path.stem}_masked{img_path.suffix}"
                        cv2.imwrite(str(output_path), masked_image)
                    except Exception as mask_error:
                        print(f"Error creating mask for {img_path}: {mask_error}")
                        output_path = masked_dir / f"{img_path.stem}_masked{img_path.suffix}"
                        cv2.imwrite(str(output_path), np.zeros_like(image))
                else:
                    print(f"No masks found for {img_path}")
                    output_path = masked_dir / f"{img_path.stem}_masked{img_path.suffix}"
                    cv2.imwrite(str(output_path), np.zeros_like(image))

            except Exception as e:
                print(f"Error processing {img_path}: {e}")
                continue

        print("\nProcessing completed!")
        print(f"Predictions saved to: {predictions_dir}")
        print(f"Masked images saved to: {masked_dir}")

        # Generate a summary
        total_processed = len(image_files)
        successful = len(list(masked_dir.glob('*')))
        print("\nProcessing Summary:")
        print(f"Total images: {total_processed}")
        print(f"Successfully processed: {successful}")
        print(f"Success rate: {(successful/total_processed)*100:.1f}%")

# === Visualization function for comparing original and processed masks ===
def visualize_results(input_dir, output_dir, model, num_samples=5):
    """
    Create a visualization comparing original images, raw model predictions,
    and final processed masks for a random sample of images.

    Parameters:
        input_dir: Directory containing input images
        output_dir: Directory containing processed results
        model: YOLO model for prediction
        num_samples: Number of random samples to visualize
    """
    input_dir = Path(input_dir)
    output_dir = Path(output_dir)
    masked_dir = output_dir / 'masked_images'

    # Create visualization directory
    vis_dir = output_dir / 'visualizations'
    vis_dir.mkdir(parents=True, exist_ok=True)

    # Get all image files
    image_files = list(input_dir.glob('*.jpg')) + list(input_dir.glob('*.png'))

    # Select random samples
    if len(image_files) <= num_samples:
        samples = image_files
    else:
        samples = random.sample(image_files, num_samples)

    print(f"Creating visualizations for {len(samples)} random samples...")

    for img_path in tqdm(samples, desc="Generating visualizations"):
        try:
            # Read original image
            original = cv2.imread(str(img_path))
            if original is None:
                continue

            # Get the processed masked image
            masked_path = masked_dir / f"{img_path.stem}_masked{img_path.suffix}"
            if masked_path.exists():
                masked = cv2.imread(str(masked_path))
            else:
                masked = np.zeros_like(original)

            # Get raw model prediction
            results = model.predict(source=original, conf=0.25)
            raw_vis = original.copy()

            # Draw raw model mask
            if len(results) > 0 and hasattr(results[0], 'masks') and results[0].masks is not None:
                masks = results[0].masks.data.cpu().numpy()
                h, w = original.shape[:2]

                # Create raw mask overlay
                raw_mask = np.zeros((h, w), dtype=bool)
                for m in masks:
                    m_resized = cv2.resize(m.astype(float), (w, h))
                    raw_mask |= (m_resized > 0.5)

                # Overlay raw mask on original image
                overlay = original.copy()
                overlay[raw_mask] = (overlay[raw_mask] * 0.7).astype(np.uint8)  # Darken the mask area
                overlay[raw_mask] += np.array([0, 0, 80], dtype=np.uint8)  # Add red tint

                # Draw contour on raw visualization
                raw_mask_uint8 = (raw_mask * 255).astype(np.uint8)
                contours, _ = cv2.findContours(raw_mask_uint8, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
                cv2.drawContours(overlay, contours, -1, (0, 255, 255), 2)  # Yellow contour

                raw_vis = overlay

            # Create comparison image
            h, w = original.shape[:2]

            # Add titles
            title_height = 40
            titled_original = np.ones((h + title_height, w, 3), dtype=np.uint8) * 255
            titled_raw = np.ones((h + title_height, w, 3), dtype=np.uint8) * 255
            titled_masked = np.ones((h + title_height, w, 3), dtype=np.uint8) * 255

            # Add images
            titled_original[title_height:, :, :] = original
            titled_raw[title_height:, :, :] = raw_vis
            titled_masked[title_height:, :, :] = masked

            # Add title text
            font = cv2.FONT_HERSHEY_SIMPLEX
            cv2.putText(titled_original, "Original Image", (10, 30), font, 0.8, (0, 0, 0), 2)
            cv2.putText(titled_raw, "Raw Model Prediction", (10, 30), font, 0.8, (0, 0, 0), 2)
            cv2.putText(titled_masked, "Processed Result", (10, 30), font, 0.8, (0, 0, 0), 2)

            # Combine images horizontally
            comparison = np.hstack((titled_original, titled_raw, titled_masked))

            # Save the comparison
            output_path = vis_dir / f"{img_name.stem}_comparison.jpg"
            cv2.imwrite(str(output_path), comparison)

        except Exception as e:
            print(f"Error creating visualization for {img_path.name}: {e}")

    print(f"Visualizations saved to: {vis_dir}")

    # Create a combined visualization of all samples
    try:
        comparison_files = list(vis_dir.glob('*_comparison.jpg'))
        if comparison_files:
            # Resize all comparison images to the same width
            target_width = 1500
            resized_images = []

            for comp_path in comparison_files:
                comp_img = cv2.imread(str(comp_path))
                if comp_img is not None:
                    h, w = comp_img.shape[:2]
                    scale = target_width / w
                    new_h = int(h * scale)
                    resized = cv2.resize(comp_img, (target_width, new_h))
                    resized_images.append(resized)

            if resized_images:
                # Stack all images vertically
                combined = np.vstack(resized_images)

                # Save the combined visualization
                combined_path = vis_dir / "all_comparisons.jpg"
                cv2.imwrite(str(combined_path), combined)
                print(f"Combined visualization saved to: {combined_path}")
    except Exception as e:
        print(f"Error creating combined visualization: {e}")

# Configuration and execution
MODEL_PATH = 'runs/segment/train/weights/best.pt'
INPUT_DIR = '/content/drive/MyDrive/placenta_backupdata- 23-3-2025/images'
OUTPUT_DIR = '/content/segmentation_results2'

try:
    print("Starting ULTRA-aggressive convex placenta segmentation...")
    processor = MaskProcessor(MODEL_PATH)

    # Configuration for ultra-convex placenta segmentation:
    processor.process_dataset(
        INPUT_DIR,
        OUTPUT_DIR,
        mask_mode='ultra_convex',      # Ultra-aggressive convexity mode
        buffer_size=5,                 # Buffer size (px)
        force_ellipse=True,            # Use ellipse fitting for perfect convexity
        pre_closing_size=50,           # Very large morphological closing (25px)
        convexity_iterations=3,        # Apply convex hull multiple times
        min_contour_area_percent=0.01, # Filter out small noise
        save_intermediate=True         # Save intermediate steps for debugging
    )

    # Generate visualizations comparing original and processed masks
    print("Creating visualizations of original vs processed masks...")
    visualize_results(INPUT_DIR, OUTPUT_DIR, processor.model, num_samples=10)

except Exception as e:
    print(f"Error: {e}")
```
